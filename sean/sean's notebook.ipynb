```python
import pandas as pd
from utils import get_Xy, preprocess, model_generator, evaluate
```

```python
%load_ext autoreload
%autoreload 2
```

```python
df = pd.read_csv('../data/predictive_maintenance.csv')
```

```python
X_train, X_test, y_train, y_test = get_Xy(df, 'Failure Type', binary=['No Failure'], drop_cols=['UDI', 'Product ID'])
```

```python
preprocessor = preprocess(X_train, y_train)
```

```python
pipeline = model_generator(X_train, y_train, preprocessor, models=None)
```

```python
evals = evaluate(pipeline, X_test, y_test)
```

```python
for e in evals:
    print(e['model'], '-' * (54 - len(e['model'])))
    print(e['report'])
```
